```python
# KNN Algorithm
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# %%
data = pd.read_csv("data.csv")

# %%
data.drop(["id","Unnamed: 32","concavity_mean"],axis=1,inplace=True)# droping datas frrom dataset->writed columns are gone!!!

data.tail()
data.head()# head of data

# malignant = M  kotu huylu tumor
# benign = B     iyi huylu tumor
```

```python
# %%
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]
# scatter plot
plt.scatter(M.radius_mean,M.texture_mean,color="yellow",label="kotu kotu",alpha= 0.8)
plt.scatter(B.radius_mean,B.texture_mean,color="black",label="iyi iyii",alpha= 0.8)
plt.xlabel("Ã§apÄ±")
plt.ylabel("gÃ¶rÃ¼ntÃ¼sÃ¼")
plt.legend()
plt.show()
```

```python
# %%
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis] # named diagnosis as 0 & 1
y = data.diagnosis.values # diagnosis values attached y
x_data = data.drop(["diagnosis"],axis=1) # axis=1 means columns to drop if its zero that means -> index to drop
```

```python
# %%
# normalization 
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)) # normalized values in 0 - 1 
```

```python
#%%
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.4,random_state=1)#splited as %60 train %40 test
```

```python
# %%
# knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print(" {} nn score: {} ".format(3,knn.score(x_test,y_test)))
```

```python
# %%
# find k value
score_list = []
for each in range(1,200):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,200),score_list) # best neighboor hood value is between 45-55
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()

score_tist = []
for each in range(30,60):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_tist.append(knn2.score(x_test,y_test))
    
plt.plot(range(30,60),score_tist) # best neighboor hood value is between 45-55
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

```python
y_pred = knn2.predict(x_test) # predict_proba difrent knn2 attributes can write here 
y_true = y_test
#%% confusion matrix
from sklearn.metrics import confusion_matrix # 
cm = confusion_matrix(y_true,y_pred)


# %% cm visualization
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize =(15,15))
sns.heatmap(cm,cmap="YlGnBu",annot = True,linewidths=5,linecolor="green",fmt = ".0f",ax=ax)#chanhed properties 
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```
